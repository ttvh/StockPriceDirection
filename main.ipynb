```python
df.set_index('Date/Time', inplace=True)
df.sort_index(inplace=True)
```

```python
df_resampled = df.resample('2min').agg({
    'Open': 'first',
    'High': 'max',
    'Low': 'min',
    'Close': 'last',
    'Volume': 'sum'
})
```

```python
df_resampled = df_resampled.dropna()
```

```python
df_resampled = df_resampled[df_resampled['Volume'] > 0]
```

```python
df_resampled.info()
```

```python
df_resampled['Next_2Min_Close'] = df_resampled['Close'].shift(-1)
```

```python
df_resampled
```

```python
df_resampled['Target_Direction'] = (df_resampled['Next_2Min_Close'] > df_resampled['Close']).astype(int)
```

```python
df_resampled
```

```python
corr_matrix = df_resampled.corr(numeric_only=True)
print(corr_matrix)

plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```

```python
# 1. Lags (Gi ng ca N*2 pht trc)
for lag in [1, 2, 3, 5, 10]:
    df_resampled[f'Close_Lag_{lag}'] = df_resampled['Close'].shift(lag)

# 2. Percent Change (Phn trm thay i so vi N*2 pht trc)
for lag in [1, 2, 3, 5, 10]:
    df_resampled[f'Pct_Change_{lag}'] = df_resampled['Close'].pct_change(lag)

# 3. Dng cc th vin TA
#  ngha window  thay i: 10 = 10 * 2 pht = 20 pht
df_resampled['SMA_10'] = SMAIndicator(df_resampled['Close'], window=10).sma_indicator()
df_resampled['SMA_50'] = SMAIndicator(df_resampled['Close'], window=50).sma_indicator() # 50*2 = 100 pht
df_resampled['RSI_14'] = RSIIndicator(df_resampled['Close'], window=14).rsi()
df_resampled['ATR_14'] = AverageTrueRange(df_resampled['High'], df_resampled['Low'], df_resampled['Close'], window=14).average_true_range()
df_resampled['MACD_diff'] = MACD(df_resampled['Close'], window_slow=26, window_fast=12, window_sign=9).macd_diff()
df_resampled['BB_%B'] = BollingerBands(df_resampled['Close'], window=20).bollinger_pband()

# 4. Features v Volume
df_resampled['Volume_SMA_20'] = SMAIndicator(df_resampled['Volume'], window=20).sma_indicator()
df_resampled['Volume_Pct_Change'] = df_resampled['Volume'].pct_change(1)
```

```python
df_model = df_resampled.dropna()
```

```python
TARGET = 'Target_Direction'
FEATURES = [col for col in df_model.columns if col not in [TARGET]]
```

```python
X = df_model[FEATURES]
y = df_model[TARGET]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2, 
    shuffle=False, 
)
```

```python
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
print(f"Kch thc Train: {len(X_train_scaled)}")
print(f"Kch thc Test: {len(X_test_scaled)}")
```

```python
def evaluate_classification(y_true, y_pred, model_name):
    print(f"\n--- Kt qu nh gi cho {model_name} ---")
    
    # 1. Accuracy
    accuracy = accuracy_score(y_true, y_pred)
    print(f"Accuracy ( chnh xc): {accuracy * 100:.2f}%")
        
    # 3. Classification Report
    print("\nClassification Report:")
    print(classification_report(y_true, y_pred, target_names=['0 (Gim)', '1 (Tng)']))
    
    # 4. Confusion Matrix
    print("Confusion Matrix:")
    ConfusionMatrixDisplay.from_predictions(y_true, y_pred, display_labels=['Gim', 'Tng'], cmap='Blues')
    plt.show()
```

```python
# 1. Khi to m hnh
# (Dng class_weight='balanced' nu y_train b mt cn bng)
model_logreg = LogisticRegression(
    random_state=42, 
    class_weight='balanced',
    max_iter=1000 # Tng max_iter  m bo hi t
)

# 2. Hun luyn
model_logreg.fit(X_train_scaled, y_train)

# 3. D on
y_pred_logreg = model_logreg.predict(X_test_scaled)

# 4. nh gi
evaluate_classification(y_test, y_pred_logreg, "Logistic Regression")
```

```python
# 1. nh ngha li tham s (param_grid)  tm kim
# (y l mt li nh  chy nhanh, bn c th thm nhiu gi tr hn)
param_grid = {
    'n_estimators': [100, 200],       # S lng cy
    'max_depth': [10, 20, None],        #  su ti a ca cy
    'min_samples_leaf': [1, 5]          # S mu ti thiu  mi l
}

# 2. Khi to GridSearchCV
# cv=3: Dng 3-fold cross-validation (chia tp train lm 3)
# n_jobs=-1: Dng tt c CPU  chy song song
grid_search_rf = GridSearchCV(
    estimator=RandomForestClassifier(random_state=42, class_weight='balanced'), # Dng class_weight nu y mt cn bng
    param_grid=param_grid,
    cv=3,
    n_jobs=-1,
    scoring='accuracy', # Tiu ch  chn m hnh tt nht
    verbose=2 # In ra tin trnh
)

# 3. Hun luyn (Tm kim) trn tp train
print("Bt u tm kim tham s tt nht...")
grid_search_rf.fit(X_train_scaled, y_train)

# 4. In ra tham s tt nht
print(f"\nTham s tt nht tm c: {grid_search_rf.best_params_}")

# 5. Ly m hnh tt nht
best_model_rf = grid_search_rf.best_estimator_

# 6. D on v nh gi
y_pred_rf = best_model_rf.predict(X_test_scaled)
evaluate_classification(y_test, y_pred_rf, "Random Forest ( tinh chnh)")
```

```python

from sklearn.utils.class_weight import compute_class_weight

# 1. To chui (Sequences)
TIME_STEPS = 30 
n_features = X_train_scaled.shape[1]

y_train_gen = y_train.values.reshape(-1, 1)
y_test_gen = y_test.values.reshape(-1, 1)

# 2. Xy dng m hnh LSTM
model_lstm = Sequential()
model_lstm.add(LSTM(units=50, activation='relu', input_shape=(TIME_STEPS, n_features), return_sequences=True))
model_lstm.add(Dropout(0.2))
model_lstm.add(LSTM(units=50, activation='relu'))
model_lstm.add(Dropout(0.2))
model_lstm.add(Dense(units=25, activation='relu'))
model_lstm.add(Dense(units=1, activation='sigmoid')) 

model_lstm.compile(
    optimizer='adam', 
    loss='binary_crossentropy', 
    metrics=['accuracy']
)

early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
print(model_lstm.summary())

# 3. To cc Generators
val_split_index = int(len(X_train_scaled) * 0.8)

# Generator cho tp Hun luyn (Train)
train_only_generator = TimeseriesGenerator(
    X_train_scaled[:val_split_index], y_train_gen[:val_split_index],
    length=TIME_STEPS, batch_size=32
)
# Generator cho tp Kim nh (Validation)
validation_generator = TimeseriesGenerator(
    X_train_scaled[val_split_index:], y_train_gen[val_split_index:],
    length=TIME_STEPS, batch_size=32
)
# *** Y L PHN B THIU ***
# Generator cho tp D on (Test) - Dng   10
test_generator = TimeseriesGenerator(
    X_test_scaled, y_test_gen, 
    length=TIME_STEPS, batch_size=1
)
# -----------------------------

# 4. Tnh ton trng s lp (class weights)
weights = compute_class_weight(
    'balanced',
    classes=np.unique(y_train),
    y=y_train
)
class_weights_dict = {cls: weight for cls, weight in zip(np.unique(y_train), weights)}
print(f"S dng Class Weights: {class_weights_dict}")

# 5. Hun luyn ( THM class_weight)
model_lstm.fit(
    train_only_generator,
    validation_data=validation_generator, 
    epochs=30, 
    callbacks=[early_stop],
    verbose=1,
    class_weight=class_weights_dict
)
```

```python
# 1. Ly y_test thc t (b ngn i TIME_STEPS)
y_test_actual = y_test_gen[TIME_STEPS:].ravel()

# 2. D on
y_pred_scaled_lstm = model_lstm.predict(test_generator)

# 3. Chuyn i xc sut (> 0.5) thnh lp (0 hoc 1)
y_pred_lstm = (y_pred_scaled_lstm > 0.5).astype(int)

# 4. nh gi
evaluate_classification(y_test_actual, y_pred_lstm, "LSTM")
```
